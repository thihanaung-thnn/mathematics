```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-2,2.1,0.1)
y = x**2 - 1
plt.plot(x,y)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlabel('$X$',fontsize=20)
plt.ylabel('$X^2$', fontsize=20)
```

```python
def gradient_descent(eta, loop):
    x = np.arange(-2,2.1,0.1)
#     plt.figure(dpi=150)
    # function y = x**2
    y = x**2 
    eta = eta
    p = 2
    plt.plot(x,y,'r--')
    
    for i in range(1,loop):
        # differentiation of f(x)
        slope_p = 2*p
        plt.plot(p, p**2,marker='o',markeredgecolor='blue',markerfacecolor='white')
        plt.text(p + 0.2, p**2,i ,color='green')
        p = p - eta*slope_p

    plt.title(f'learning rate - {eta}\nlooping - {loop} times')
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    
    plt.xlabel('$X$',fontsize=20)
    plt.ylabel('$X^2$', fontsize=20)
```

```python
gradient_descent(0.1,20)
```

```python
gradient_descent(0.02,20)
```

```python
gradient_descent(0.9,20)
```

```python
gradient_descent(1,100)
```

```python
def gradient_descent_two(eta, loop, start_point):
#     plt.figure(dpi=150)
    x = np.arange(-2,2.1,0.1)
    # function y = x**2
    y = x**4 - 4*x**2 + x
    eta = eta
    p = start_point
    points = []
    
    for i in range(loop):
        # differentiation of f(x)
        slope_p = 4*p**3 - 8*p + 1
        plt.plot(p, p**4 - 4*p**2 + p,marker='o',markeredgecolor='blue',markerfacecolor='white')
        plt.text(p + 0.2, p**4 - 4*p**2 + p,i ,color='green')
        p = p - eta*slope_p
        points.append(p)
#     print(points)
    plt.plot(x,y,'r--')
    plt.title(f'learning rate - {eta}\nstart point - {start_point}')
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    
    plt.xlabel('$X$',fontsize=20)
    plt.ylabel('$X^4 - 4X^2 + X$', fontsize=20)
```

```python
gradient_descent_two(0.1,10,2)
```

```python
gradient_descent_two(0.1,10,-2)
```

```python
def gradient_descent_three(eta, loop):
    X = np.arange(-5, 5, 0.25)
    Y = np.arange(-5, 5, 0.25)
    X, Y = np.meshgrid(X, Y)
    Z = X**2 + Y**2
#     plt.figure(dpi=150)
    fig = plt.figure(figsize = (10,7))
    contours = plt.contour(X, Y, Z, 30)

    plt.xlabel('$X^2$', fontsize=20)
    plt.ylabel('$Y^2$', fontsize=20)

    pointx,pointy = 4,4
    learning_rate = eta
    for i in range(1,loop):
        pointx = pointx - learning_rate*2*pointx
        pointy = pointy - learning_rate*2*pointy
        plt.plot(pointx, pointy,'bo')
        plt.text(pointx+0.3,pointy,i)
    plt.colorbar()
    plt.title(f"Gradient descent of function $x^2 + y^2$\nlearning rate = {learning_rate}", fontsize=15)
    
gradient_descent_three(0.1,30)
```
